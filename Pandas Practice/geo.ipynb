```python
from geopy.geocoders import ArcGIS
import pandas as pd
nom = ArcGIS()
```

```python
df = pd.read_csv("supermarkets.csv")
```

```python
df["Address"] = df["Address"] + ", " + df["City"] + ", "  + df["State"] + ", " + df["Country"]
df["Coordinantes"] = df['Address'].apply(nom.geocode)
df["Latitude"] = df["Coordinantes"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Coordinantes"].apply(lambda x: x.longitude if x != None else None)
df = df.drop("Coordinantes", axis = 1)
df
```
